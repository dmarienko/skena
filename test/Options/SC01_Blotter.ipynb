### <font color='orange'>2.5 Remarkable Properties of BM</font>

#### <font color='green'>Property 2.1</font> : Derivative of $W_t$

We know $W_t \sim N(0,t)$, so $\mathbb{E}[W_t^2] = t$. 

If you tried to compute the "derivative" as: 
$$
\lim_{\Delta t \to 0} \frac{W_{t+\Delta t} - W_t}{\Delta t}
$$                  

Question: What happens to the variance of this ratio as $\Delta t \to 0$



1. $\mathbb{E}[W_t^2] = Var(W_t) = t$ because of this :: variation: $Var(Y) = E[(Y - E[Y])^2] = E[Y^2]$ 


2. $W(t+\Delta t) - W_n(t) \stackrel{d}= W(\Delta t)$ because of additive property of BM process  
so variation of
$$
    Var\left(\lim_{\Delta t \to 0} \frac{W_{t+\Delta t} - W_t}{\Delta t}\right) 
    = Var \left( \lim_{\Delta t \to 0} \frac{W(\Delta t)}{\Delta t}\right) 
    = \lim_{\Delta t \to 0} Var \left( \frac{W(\Delta t)}{\Delta t}\right)
    = \lim_{\Delta t \to 0} \frac{1}{\Delta t^2} \Delta t = \infty 

$$


> The "derivative" has infinite variance - it doesn't converge to anything! This is why Brownian motion is nowhere differentiable.

>Physical intuition: Brownian motion is "too rough" - it wiggles so violently at small scales that any attempt to measure a slope produces wild, unbounded fluctuations.  

#### <font color='green'>Property 2.2</font> : Unbounded range (hits any value eventually)

Statement: Brownian motion will eventually hit any real value, no matter how large or how negative. Formally, for any $a \in \mathbb{R}$: 
$$
  \mathbb{P}\left(\sup_{t \geq 0} W_t \geq a\right) = 1 \quad \text{and} \quad  
  \mathbb{P}\left(\inf_{t \geq 0} W_t \leq a\right) = 1
$$ 

Even more striking: if $W_t = 1{,}000{,}000$ right now, it will (with probability 1) return to zero at some later time.                 

**Intuition**: At time $t$, we have $W_t \sim N(0, t)$. As $t \to \infty$, the standard deviation $\sqrt{t} \to \infty$. So the distribution "spreads out" without bound.
More precisely, for any level $a > 0$: 
$$\mathbb{P}(W_t > a) = \mathbb{P}\left(\frac{W_t}{\sqrt{t}} > \frac{a}{\sqrt{t}}\right) = 1 - \Phi\left(\frac{a}{\sqrt{t}}\right)$$
As $t \to \infty$, we have $\frac{a}{\sqrt{t}} \to 0$, so:
$$
    \mathbb{P}(W_t > a) \to 1 - \Phi(0) = \frac{1}{2}
$$
This means at large times, there's roughly 50% chance of being above any fixed level!
